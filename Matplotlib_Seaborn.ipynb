## Adjust Seaborn Legend

```python
iris = sns.load_dataset("iris")
sns.set_style("darkgrid")
sns.color_palette()

plot = sns.displot(iris, x="petal_width", hue="species")

# Change Axis names
plt.xlabel("Petal Width (cm)")
plt.ylabel("Count")

# Change legend properties
legend = plot._legend

legend.set_title("Species")
for t, l in zip(legend.texts,('setosa', 'versicolor', 'virginica')):
    t.set_text(l)

# Adjust legend location
plot.fig.subplots_adjust(top=0.9,right=0.8)
```

## Adjust Seaborn Legend: Matplotlib

```python
iris = sns.load_dataset("iris")
sns.set_style("darkgrid")
sns.color_palette()

plot = sns.displot(iris, x="petal_width", hue="species", legend=False)
plt.xlabel("Petal Width (cm)")
plt.ylabel("Count")

plt.legend(title='Species', loc='best', labels=['setosa', 'versicolor', 'virginica'])
plt.show(plot)
```

## Specify Graph Colors

```python
iris = sns.load_dataset("iris")
sns.set_style("darkgrid")
sns.color_palette()

plot = sns.displot(iris, x="petal_width", hue="species", palette=sns.color_palette(palette = ["SteelBlue" , "Salmon", "plum"]))

# Change Axis names
plt.xlabel("Petal Width (cm)")
plt.ylabel("Count")

# Change legend properties
legend = plot._legend

legend.set_title("Species")
for t, l in zip(legend.texts,('setosa', 'versicolor', 'virginica')):
    t.set_text(l)

plot.fig.subplots_adjust(top=0.9,right=0.8)
```

```python
import matplotlib.pyplot as plt
import random

# x and y coordinates
x_coordinates = [range(1,4)]*2
y_coordinates = [[3,4,5],[2,2,2]]

# colors (one per series --> 2 in this example)
colors = []
r = random.random
for i in range(0,len(x_coordinates)):
    rgb =  (r(),r(),r())
    colors.append(rgb)

# iterate over the series
for x, y, c in zip(x_coordinates, y_coordinates, colors):
    plt.plot(x, y, color=c, linestyle='-', marker='o', markeredgecolor='k',
            markersize=10)

# show the figure
plt.margins(x=0.1, y=0.1)
plt.show()
```

### Line Graph: add Legend

```python
import numpy as np
import matplotlib.pyplot as plt

# Make some fake data.
a = b = np.arange(0, 3, .02)
c = np.exp(a)
d = c[::-1]

# Create plots with pre-defined labels.
fig, ax = plt.subplots()
ax.plot(a, c, 'k--', label='Model length')
ax.plot(a, d, 'k:', label='Data length')
ax.plot(a, c + d, 'k', label='Total message length')

legend = ax.legend(loc='upper center', shadow=True, fontsize='large')

# Put a nicer background color on the legend.
legend.get_frame().set_facecolor('C8')
# ax.patch.set_alpha(0.5)

plt.show()
```

## SciPy Stats

### Pearson Correlation Coefficient: .pearsonr(x, y)

<ins>Format:</ins> scipy.stats.pearsonr(x, y)

Pearson correlation coefficient and p-value for testing non-correlation. Varies between -1 and +1 with 0 implying no correlation. Correlations of -1 or +1 imply an exact linear relationship. Positive correlations imply that as x increases, so does y. Negative correlations imply that as x increases, y decreases.

For more: <a href="https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.pearsonr.html">Stats documentation</a>

```python
r, p = stats.pearsonr(Mcdonalds_df['Protein'], Mcdonalds_df['Total Fat'])

print(r, p)
```

## Joint Scatterplot

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import scipy.stats as stats

plot = sns.jointplot(x='Protein', y='Total Fat', data=Mcdonalds_df)
```

### Joint Scatterplot: Add Legend

```python
# if you choose to write your own legend, then you should adjust the properties:
# phantom = plot.ax_joint.plot([], [], linestyle="", alpha=0)
# here graph is not a ax but a joint grid, so we access the axis through ax_joint method

# Rounding notation for floats (two decimals is {:.2f}, format for order of display

plot = sns.jointplot(x='Protein', y='Total Fat', data=Mcdonalds_df)

r, p = stats.pearsonr(Mcdonalds_df['Protein'], Mcdonalds_df['Total Fat'])
plot.ax_joint.legend(['r={:.2f}, p={:.2e}'.format(r,p)], loc="upper left")

```

### Joint Scatterplot: Add Linear Regression fit line (uses regplot)

```python
# jointplot() can use regplot() to show the linear regression fit on the joint axes by passing kind="reg"

plot = sns.jointplot(x='Protein', y='Total Fat', data=Mcdonalds_df, kind="reg")

r, p = stats.pearsonr(Mcdonalds_df['Protein'], Mcdonalds_df['Total Fat'])
plot.ax_joint.legend(['r={:.2f}, p={:.2e}'.format(r,p)], loc="upper left")

```
